### Ejercicio
* Inicilamente, vamos a copiar un subconjunto de datos correspondiente al dataset para operar con l que denominaremos *subdata*
* Vamos a eliminar los valores duplicados mediante la funcin *duplicate*, de hecho se eliminaran los datos duplicados que contengan idnticos nombre, localizacin fecha de apertura que denominaremos *sudata_clean*

````python
import pandas as pd

data = pd.read_csv('data/coaster_db.csv')

subdata = data[['coaster_name',
    'Location', 'Status',
    'Manufacturer',
    'year_introduced',
    'latitude', 'longitude',
    'Type_Main',
    'opening_date_clean',
    'speed_mph', 
    'height_ft',
    'Inversions_clean', 'Gforce_clean']].copy()
````

````python
# Datos duplicados
data[data.duplicated(subset=['coaster_name'])]
````

````python
# Verificar que son datos duplicados, por ejemplo la montaa rusa con nombre 'Derby Racer'
subdata[subdata['coaster_name']=='Derby Racer']
````

````python
# Eliminar los datos duplicados que tienen mismo nombre, localizacin fecha de apertura
subdata_clean = subdata.loc[~subdata.duplicated(subset=['coaster_name','Location','opening_date_clean'])].reset_index(drop=True).copy()
````

````python
subdata_clean['speed_kmh'] = subdata_clean['speed_mph']*1.60934
subdata_clean['height_m']  = subdata_clean['height_ft']*0.3048
````

````python
# Histograma
import numpy as np 
hist_data = subdata_clean['year_introduced'].value_counts()

h = pd.DataFrame(data=hist_data)
h = h.sort_index()
fig, ax = plt.subplots(figsize=(10,5))
fig.suptitle='Evolution of coaster installed'


ax.bar(h.index, h['year_introduced'], label='coaster')


ax.set_xticks(range(h.index[0], h.index[-1], 10))
nuevas_etiquetas=np.arange(h.index[0], h.index[-1], 10)

ax.set_xticklabels(nuevas_etiquetas,  rotation=45)
````
